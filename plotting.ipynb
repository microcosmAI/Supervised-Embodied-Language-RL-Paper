```python
import seaborn as sns
import pandas as pd
from statistics import mean 
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
data = pd.read_csv("/home/cowolff/Documents/GitHub/s.language_experiments/PlottingData/Accuracy PPO Sender, lr=1e-06, batch=16, envs=10, Window Size 3_10.csv")

y = [mean(data["Value"][x:x+20])*100 for x in range(len(data["Value"])-20)]
x = [data["Step"][x] for x in range(len(data["Step"])-20)]

# Plot the responses for different events and regions
fig, ax = plt.subplots()
p = sns.lineplot(x=x, y=y)
p.set(xlabel="Timestep", ylabel="Accuracy (%)")
ax.set_ylim(20, 95)
plt.savefig("Plots/AccuracySender.pdf", format='pdf')
```
