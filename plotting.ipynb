```python
import seaborn as sns
import pandas as pd
from statistics import mean 
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
data = pd.read_csv("PlottingData/Accuracy PPO Test, lr=1e-06, size=64, batch=96, envs=10_1.csv")

y = [mean(data["Value"][x:x+20])*100 for x in range(len(data["Value"])-20)]
x = [data["Step"][x] for x in range(len(data["Step"])-20)]

# Plot the responses for different events and regions
fig, ax = plt.subplots()
p = sns.lineplot(x=x, y=y)
p.set(xlabel="Timestep", ylabel="Accuracy (%)")
ax.set_ylim(50, 85)
plt.savefig("Plots/AccuracyReceiver.pdf", format='pdf')
```

```python
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
data = pd.read_csv("PlottingData/Variance PPO Test, lr=1e-06, size=64, batch=96, envs=10_1.csv")

y = [mean(data["Value"][x:x+20]) for x in range(len(data["Value"])-20)]
x = [data["Step"][x] for x in range(len(data["Step"])-20)]

# Plot the responses for different events and regions
fig, ax = plt.subplots()
p = sns.lineplot(x=x, y=y)
p.set(xlabel="Timestep", ylabel="Variance")
ax.set_ylim(0.2, 1)
plt.savefig("Plots/VarianceReceiver.pdf", format='pdf')
```
